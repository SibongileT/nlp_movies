### Scrape IMDb for movies above a 6.5 rating between 1950-2020

```python
from bs4 import BeautifulSoup
import json
import urllib
from time import sleep
import pandas as pd
import pickle
import requests
from random import randint
from time import sleep
import pandas as pd

def parse(url):
    '''
    Get scrape a page for it's title, titleid, and genre
    '''
    df = pd.DataFrame()
    movie_list = []
    id_list = []
    genre_list = []
    response = requests.get(url)
    print(url)
    page = response.text
    soup = BeautifulSoup(page, "lxml")
    bs = BeautifulSoup(page, 'lxml')
    movies = bs.find_all('div', attrs={'class':'lister-item-content'})
    for movie in movies:
        title = movie.find('a').get_text()
        href = movie.find('a').attrs['href']
        synopsis = movie.find_all('p')[1]
        link = synopsis.find('a')
        genre = movie.find('span', attrs={'class':'genre'})
        if genre:
            genre = genre.get_text().strip()
        else:
            genre = ''
        if href[7:17][9] == '/':
            title_id = href[7:16]
        else:
            title_id = href[7:17]
        movie_list.append(title)
        id_list.append(title_id)
        genre_list.append(genre)
    sleep(randint(2,10))
    return movie_list,id_list,genre_list

movie_info_df = pd.DataFrame(columns=['title','title_id','genre_list'])
movies_list = []
ids_list = []
genres_list = []
temp_df = pd.DataFrame(columns=['title','title_id','genre_list'])
for i in range(1950,2021):
    print("Year is: " + str(i))
    year  = i
    j = 1
    while(True):
        if j == 1:
            movie_list,id_list,genre_list = parse('https://www.imdb.com/search/title/?title_type=feature&release_date='+str(i)+'-01-01,'+str(i)+'-12-31&user_rating=6.5,')
            movies_list.extend(movie_list)
            ids_list.extend(id_list)
            genres_list.extend(genre_list)
            #year_list.extend(str(year))
        else:
            movie_list,id_list,genre_list = parse('https://www.imdb.com/search/title/?title_type=feature&release_date='+str(i)+'-01-01,'+str(i)+'-12-31&user_rating=6.5,&start='+str(j)+'&ref_=adv_nxt')
            movies_list.extend(movie_list)
            ids_list.extend(id_list)
            genres_list.extend(genre_list)
            # If movie list is empty that means there are no more pages and want to move onto the next year
            if not movie_list:
                if not movies_list:
                    break      
                # If movie list is empty but there are still movies in the movie list pickle that information and then move onto the next year
                else:
                    temp_df = pd.DataFrame(columns=['title','title_id','genres'])
                    temp_df['title'] = movies_list
                    temp_df['title_id'] = ids_list
                    temp_df['genres'] = genres_list
                    movies_list=[]
                    ids_list=[] 
                    genres_list = []
                    movie_info_df = movie_info_df.append(temp_df,ignore_index = True)
                    movie_info_df.to_pickle("final_movie_info.pkl")
                    break
        # After scraping 8 pages pickle object
        if j == 401:
            temp_df = pd.DataFrame(columns=['title','title_id','genres'])
            temp_df['title'] = movies_list
            temp_df['title_id'] = ids_list
            temp_df['genres'] = genres_list
            movies_list=[]
            ids_list=[]
            genres_list = []
            movie_info_df = movie_info_df.append(temp_df,ignore_index = True)
            movie_info_df.to_pickle("final_movie_info.pkl")
        j+=50
        sleep(randint(2,10))  
        
# except:
#      print("Error!")
```
